```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample1 = np.loadtxt("./radioactive_sample_1.txt")
```

```python
averages = np.zeros(len(sample1))
cumsum = 0
for i in range(len(sample1)):
    cumsum += sample1[i]
    averages[i] = cumsum / (i+1)
```

```python
plt.plot(range(1, len(sample1)+1), averages)
```

```python
# estimate for c
c = averages[-1]
c
```
